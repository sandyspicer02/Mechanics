```python
from scipy.stats import norm
import numpy as np
import matplotlib.pylab as plt
%matplotlib notebook
```

```python
# Assign number to color paper

p = 1
y = 2
b = 3
g = 4
w = 5
k = 6
```

We are going to note which color paper we pick out of the cup first for 40 different measurements

```python
# Truncated Gaussian

values = np.array([b+k,y+b,y+g,w+w,k+g,w+w,y+w,b+w,w+p,w+k,k+g,k+y,g+p,b+g,w+g,p+b,g+w,b+b,k+y,g+k,g+y,g+w,w+p,p+p,w+b,k+g,y+w,w+p,w+k,p+b,k+w,k+y,b+k,g+k,g+k,y+k,k+k,y+w,w+y,w+p,g+w,y+b,b+p,b+w,p+p,w+k,k+k,w+g,y+w,g+y,g+w,g+k,p+p,b+p,y+y,w+y,w+g,w+g,k+k,k+p])
print values
print np.mean(values)
```

```python
def gauss(x, A, mean, std):
    g = A*np.exp(-(x-mean)**2/(2*std**2))
    return g
```

```python
# Plots the histogram
plt.figure(figsize=(8,6))
plt.hist(values,bins=11) # 11 different combinations (11 bins)

xfit = np.linspace(0,14,100)
y = gauss(xfit,10,8.5,3)
plt.plot(xfit,y)

plt.title('Likelihood of Color Match-up')
plt.xlabel('Sum')
plt.ylabel('Entries')
plt.savefig('color-matchup')
```

```python
# Calculates the Gaussian distribution
mean,std = norm.fit(values)
xmin = min(values)
xmax = max(values)

x = np.linspace(xmin,xmax,100)
y = norm.pdf(x,mean,std)
plt.plot(x,y,'r-')
```
