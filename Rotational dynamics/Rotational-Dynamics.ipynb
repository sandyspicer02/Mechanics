## Pulley 1 - 100g

```python
time1, angle1, velocity1, acceleration1 = np.loadtxt('1_100g.txt', unpack=True, skiprows=7, usecols=[0,1,2,3])
time2, angle2, velocity2, acceleration2 = np.loadtxt('2_100g.txt', unpack=True, skiprows=7, usecols=[0,1,2,3])
time3, angle3, velocity3, acceleration3 = np.loadtxt('3_100g.txt', unpack=True, skiprows=7, usecols=[0,1,2,3])

plt.plot(time1,acceleration1,label='Trial 1')
#plt.plot(time2,acceleration2,label='Trial 2')
#plt.plot(time3,acceleration3,label='Trial 3')
plt.xlabel('Time (s)')
plt.ylabel('Angular Acceleration (rad/s^2)')
plt.title('Angular Acceleration vs. Time')
plt.legend()
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('AVT_p1_100g.png')
plt.xlim(0,5)
plt.ylim(0,3)
```

```python
good1 = (time1>0) * (time1<4)
meanal_1 = np.mean(acceleration1[good1])
sigal_1 = np.std(acceleration1[good1]) / np.sqrt(np.count_nonzero(good1)-1)
print meanal_1, sigal_1

tor1 = 0.005 * 0.15 * 9.8  # torque = rmg
print(tor1)

logalpha1 = np.log10(meanal_1)
logalphaerr1 = sigal_1 / meanal_1 / np.log(10)
logtorque1 = np.log10(0.005 * 0.15 * 9.8)

fig, ax = plt.subplots()
ax.scatter(time1, acceleration1, s = 50)
ax.scatter(time1[good1], acceleration1[good1],c='r',s=50)
ax.set_ylim(-5,8)
ax.set_ylabel('Angular Acceleration (rad/s^2)')
ax.set_xlabel('Time (s)')
plt.title('Pulley 1: Angular Acceleration vs. Time 100g')
plt.legend()
plt.savefig('pulley1_full_accel_vs_t_100g')
```
