```python
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
```

```python
r = np.linspace(0,10,1000)
A=1
s=1
R=1
U = A*((np.exp((R-r)/s)-1)**2-1)

plt.plot(r,U,'r-')
plt.xlabel('Distance Between Atoms (m)')
plt.ylabel('Potential Energy (J)')
plt.title('Potential Energy vs. Distance Between Atoms')
plt.savefig('U_vs_r')
```
